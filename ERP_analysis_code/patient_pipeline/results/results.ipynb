```python
static = 0
counter = 3
p1_performances_updated = dict()

for k in performances_new.keys():
    if (static%2==0):
        p1_performances_updated.update({f"p1_s{counter}_static": performances_new.get(k)})
        static += 1
    else:
        p1_performances_updated.update({f"p1_s{counter}_adaptive_sw": performances_new.get(k)})
        counter +=1
        static +=1    

print(performances_new.keys())
print(p1_performances_updated.keys())
for k,i in zip(performances_new.keys(), p1_performances_updated.keys()):
    print(performances_new.get(k) == p1_performances_updated.get(i))

```
